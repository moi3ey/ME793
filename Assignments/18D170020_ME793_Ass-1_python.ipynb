```python
import numpy as np              
from time import process_time                        #to calculate time
from tabulate import tabulate                        #to make table

my_table = []
for i in range(5,11):                                #looping for n= 5 to n=10      
  time_start = process_time()                        #starts noting the time   
  MatX = np.random.randint(8,size=(i,i))             #random matrix generated
  print(MatX)   
  detX = np.linalg.det(MatX)                      #checking for singular matrix
  
  if detX!=0:                                     #print detX
    invX=np.linalg.inv(MatX)                      #inverse of matrix calculated
    print(invX)
  else:
    print("Singular Matrix, no inverse found")
  time_stop = process_time()                      #time count stops
  net_time = time_stop-time_start                 #net = time stop - time start
  print("net time for n=%s = "%i ,net_time ,  "seconds")
  frobeniusnorm = np.linalg.norm(invX)
  entrydata= ("n= %s"%i, net_time, frobeniusnorm)           #for table purpose
  
  my_table.append(entrydata)                                #print entrydata

tableX = np.array(my_table)                                 #print tableX
                  

head = ["n" , "time" , "frobenius norm"]
XvsY = tabulate(tableX, headers = head, tablefmt="grid")
print(XvsY)
```
