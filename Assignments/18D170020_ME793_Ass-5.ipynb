Task 3- Now divide the
complete data into 7 subgroups of thermal conductivity. Suppress the groups 2, 4 and 6. 

```python
#Other properties removed from this table
df0 = df[['Thermal_Conductivity']]
df0
```

```python
#Group 1 shuffled
df1 = df0.iloc[0:11]
x1 = df1.sample(frac=1)
x1
```

```python
#Group 3 shuffled
df3 = df0.iloc[21:37]
x3 = df3.sample(frac=1)
x3
```

```python
#Group 5 shuffled
df5 = df0.iloc[51:63]
x5 = df5.sample(frac=1)
x5
```

```python
#Group 7 shuffled
df7 = df0.iloc[77:79]
x7 = df7.sample(frac=1)
x7
```

Task 4- Apply k-Means method and show that there are 4 groups of elements.

```python
X = np.concatenate((x1,x3,x5,x7),axis=0)
plt.plot(X[:,0],'k.')
#plt.plot(X[:,0],X[:,1],'k.')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
n = 4
k_means = KMeans(n_clusters=n)
k_means.fit(X)
```

```python
centroids = k_means.cluster_centers_
labels= k_means.labels_
```

```python
plt.plot(X[labels==0,0],'r.', label='cluster 1')
plt.plot(X[labels==1,0],'b.', label='cluster 2')
plt.plot(X[labels==2,0],'g.', label='cluster 3')
plt.plot(X[labels==3,0],'k.', label='cluster 4')
plt.plot(centroids[:,0],'mo',markersize=8, label='centroids')

plt.legend(loc='best')
plt.show()
```

So,  there are 4 groups of elements.
